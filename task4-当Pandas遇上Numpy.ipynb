### 83从Numpy数组创建DataFrame

-  np.arange()函数分为一个参数，两个参数，三个参数三种情况
    - 1）一个参数时，参数值为终点，起点取默认值0，步长取默认值1。
    - 2）两个参数时，第一个参数为起点，第二个参数为终点，步长取默认值1。
    - 3）三个参数时，第一个参数为起点，第二个参数为终点，第三个参数为步长。其中步长支持小数

```python
#使用numpy生成20个0-100固定步长的数
data = np.arange(0, 100, 5)
df2 = pd.DataFrame(data)
df2
```

### 84. 从Numpy数组创建DataFrame

- numpy.random.normal(loc=0,scale=1e-2,size=shape) 
    - loc(float)：正态分布的均值，对应着这个分布的中心。loc=0说明这一个以Y轴为对称轴的正态分布，
    - scale(float)：正态分布的标准差，对应分布的宽度，scale越大，正态分布的曲线越矮胖，scale越小，曲线越高瘦。
    - size(int或者整数元组)：输出的值赋在shape里，默认为None。 

```python
# 使用numpy生成20个指定分布(如标准正态分布)的数
tem = np.random.normal(0, 1, 20)
df3 = pd.DataFrame(tem)
df3
```

### 89. 提取第一列中不在第二列出现的数字

- isin()
    - 判断数据是否存在，返回布尔型向量
    - 可用于数据的过滤
    
    - Series.isin(vals):判断Series的内容是否再vals中
    - Index.isin(vals):判断索引值是否在vals中
    - DataFrame.isin(vals)
        - vals为列表时，判断DataFrame中数据是否在vals中
        - vals为字典时，判断DataFrame中对应列的数据是否在vals中
    - https://blog.csdn.net/lildn/article/details/114703898

```python
df['col1'][~df['col1'].isin(df['col2'])] #~取反
```

### 97. 对第二列计算移动平均值

```python
# 每次移动三个位置，不可以使用自定义函数
#方法一：
df['col2'].rolling(3).mean()
```

```python
#方法二：
np.convolve(df['col2'], np.ones(3)/3, mode='valid')
```

- numpy.convolve(a, v, mode='full'),这是numpy函数中的卷积函数库
    - 参数：
        - a:(N,)输入的一维数组
        - b:(M,)输入的第二个一维数组
        - mode:{‘full’, ‘valid’, ‘same’}参数可选
            - ‘full’　默认值，返回每一个卷积值，长度是N+M-1,在卷积的边缘处，信号不重叠，存在边际效应。
            - ‘same’　返回的数组长度为max(M, N),边际效应依旧存在。
            - ‘valid’ 　返回的数组长度为max(M,N)-min(M,N)+1,此时返回的是完全重叠的点。边缘的点无效。
    - https://blog.csdn.net/u011599639/article/details/76254442

### 100. 计算第二列与第三列之间的欧氏距离

```python
#方法一：(公式法)
import math as mt
mt.sqrt(sum((df['col2']-df['col3'])**2))
```

```python
#方法二：
np.linalg.norm(df['col2'] - df['col3'])
```

- numpy.linalg.norm(求范数)
    - linalg=linear（线性）+algebra（代数），norm则表示范数。
    - x_norm=np.linalg.norm(x, ord=None, axis=None, keepdims=False)  　
        - ①x: 表示矩阵（也可以是一维）
        - ②ord：范数类型
        - ③axis：处理类型
        - ④keepding：是否保持矩阵的二维特性
    - https://www.cnblogs.com/zongfa/p/8745853.html
